```python
from sklearn.datasets import make_moons, make_circles
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import svm
import numpy as np
```

```python
X,y = make_moons(noise=0.3, random_state=0)

print(X.shape)
print(y.shape)
print(X)
print(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="plasma")
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.show()
```

```python
classifier = svm.SVC(kernel='linear', C=10.0)
classifier.fit(X, y)
```

```python
# pas de la grille
h = .02

# coordonnes min et max suivant les 2 coordonnes
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

# cration d'une grille (maillage) suivant les coordonnes
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))

# Z est la fonction de dcision
Z = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
# redimensionnement pour l'affichage
Z = Z.reshape(xx.shape)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap="plasma")
plt.xlabel("x_1")
plt.ylabel("x_2")

# trac sous forme d'un contour
plt.contour(xx, yy, Z, colors='red', levels=[0], linestyles=['-'])
plt.show()
```

```python
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.contourf(xx, yy, Z, cmap=plt.cm.hot, alpha=.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="YlGnBu")

plt.show()
```

```python
X_test = np.array([[1.47, -0.99],[-0.55, 0.78],[2.25,-1.25]])
```

```python
y_test = classifier.predict(X_test)
# affichage du rsultat
print(y_test)
```

```python
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.contourf(xx, yy, Z, cmap=plt.cm.hot, alpha=.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="YlGnBu")
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap="PiYG")
plt.show()
```
